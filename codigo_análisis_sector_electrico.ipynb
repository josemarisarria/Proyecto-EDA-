# EMPEZAMOS A TRABAJAR EL SIGUIENTE DATA FRAME 
### EN ESTA OCASION EL TRABAJO SERA MAS RAPIDO YA QUE TENEMOS LA FUNCION GRAFICA QUE LA PODEMOS APROVECHAR
##### MISMA ESTRATEGIA QUE ANTES, VER LOS DATOS, RENOMBRAR COLUMNAS, BORRAR DATOS y/o RENOMBRAR NAN 
###### ELEGIMOS CON UNA MASCARA LOS VALORES QUE QUEREMOS EN UNA NUEVA VARIABLE 

```python
###     SIGUIENTE DATAFRAME     #####

###### DENTRO DE LAS ENERGIAS RENOVABLES, VAMOS A MIRAR LAS QUE MAS SE INSTALAN A NIVEL MUNDIAL Y VEAMOS SU PRODUCCION Y CONSUMO

solar=pd.read_csv('./data/generacion solar.csv')
solar.info()

### ADAPTAMOS LA TABLA

solar.rename(columns={"Electricity from solar (TWh)": "CONSUMO SOLAR (TWh)"}, inplace=True)
solar.rename(columns={"Entity": "Pais"}, inplace=True)
solar.rename(columns={"Year": "Año"}, inplace= True)
solar.drop('Code', axis=1, inplace=True) 

#### VEMOS QUE MUCHOS DATOS SON INAPRECIABLES, SOBRETODO HASTA UN CIERTO AÑO, YA QUE LAS INSTALACIONES SOLARES NO EMPEZARON HASTA ACERCARSE LOS NOVENTA O 2000 
### MODIFICAMOS EL DATA FRAME QUITANDO AÑOS 

solar=solar[solar['Año']>1995]
solar

display(solar)
solar.describe()

```

##### MIRAMOS PAISES AGRUPADOS CON SUS CONSUMOS POR VER SI TENDREMOS DATOS INTERESANTES

```python

display(solar.groupby('Pais')['CONSUMO SOLAR (TWh)'].value_counts())
### segun pais, la instalacion de solar es inapreciable
display(solar['Pais'].unique())
```

#### SACAMOS LOS DATOS DE LA UNION EUROPEA

### EMPEZAMOS A SACAR GRAFICAS

```python
##### APROVECHO LA FUNCION DE GRAFICA CREADA ARRIBA PARA TRABAJAR ESTE DATA FRAME
### COPIO LA FUNCION DE ARRIBA Y LO ADAPTO AL NOMBRE DEL NUEVO DATAFRAME
### Así puedo graficar cualquier país, la evolución de renovables

def grafica(pais, tipo):
    AAA=solar['Pais']==pais
    datos_AAA=solar[AAA]
    x=datos_AAA['Año']
    y=datos_AAA['CONSUMO SOLAR (TWh)']

    if tipo == plt.plot: 
        plt.plot(x, y)
        plt.title("Consumo solar", color= 'red')
        plt.xlabel("AÑO", color='red')
        plt.ylabel("TWh", color='red');
        
    elif tipo == plt.bar: 
        plt.bar(x, y)
        plt.title("Consumo solar", color= 'red')
        plt.xlabel("AÑO", color='red')
        plt.ylabel("TWh", color='red');

grafica('European Union (27)', plt.plot)
```

```python
#### SACO ALGUNAS GRAFICAS MÁS PARA VISUALIZAR

## EVOLUCION MUNDIAL 

grafica('World', plt.plot)
```

```python
#### ARRIBA UTILIZO LA FUNCION QUE HE HECHO PARA SACAR GRAFICAS EN UN SEGUNDO 
plt.figure(figsize=(7,10))#Se  inicializa la figura

plt.subplot(3, 1, 1) # filas, columnas, posición
grafica('World', plt.plot)
plt.title("CONSUMO SOLAR A NIVEL MUNDIAL", color= 'blue')
plt.xlabel("", color='red')
plt.ylabel("TWh", color='red');

plt.subplot(3, 1, 2) # filas, columnas, posición
grafica('European Union (27)', plt.plot)
grafica('Spain',plt.bar)
plt.title("CONSUMO SOLAR EN EUROPA   /    España en columnas", color= 'green')
plt.xlabel("", color='red')
plt.ylabel("TWh", color='red');

plt.subplot(3,1,3)
grafica('Africa',plt.plot)
grafica('South Africa', plt.plot)
plt.title("CONSUMO SOLAR EN AFRICA / SudAfrica naranja", color= 'black')
plt.xlabel("", color='red')
plt.ylabel("TWh", color='red');



```

## EMPEZAMOS A TRABAJAR EL DATAFRAME CON LOS DATOS DE ENERGIA EOLICA **MISMA ESTRATEGIA QUE ARRIBA**

```python
###### VAMOS A PONER LOS DATOS DE ENERGÍA EÓLICA, OTRA DE LAS FUENTES DE GENERACIÓN DE ENERGIA MAS FUERTE

eolico=pd.read_csv('./data/generacion eolica.csv')
eolico.info()
```

```python
##### PREPARAMOS LA TABLA PARA TRABAJAR


### ADAPTAMOS LA TABLA

eolico.rename(columns={"Electricity from wind (TWh)": "TWh eolico"}, inplace=True)
eolico.rename(columns={"Entity": "Pais"}, inplace=True)
eolico.rename(columns={"Year": "Año"}, inplace= True)
eolico.drop('Code', axis=1, inplace=True) 
```

```python
### AL IGUAL QUE EN LAS OTRAS TABLAS, HAY MUCHOS AÑOS CON VALORES DESPRECIABLES

eolico=eolico[eolico['Año']>1995]
display(eolico)

```

### EMPEZAMOS A GRAFICAR, USAMOS EL FORMATO DE FUNCION PARA QUE SEAN GRAFICAS INMEDIATAS

```python
#### EMPEZAMOS A GRAFICAR, AL IGUAL QUE ANTES A TRAVES DE UNA FUNCION Y SE HACE RAPIDO 

def grafica_eolica(pais, tipo):
    AAA=eolico['Pais']==pais
    datos_AAA=eolico[AAA]
    x=datos_AAA['Año']
    y=datos_AAA['TWh eolico']

    if tipo == plt.plot: 
        plt.plot(x, y)
        plt.title("Consumo eolico", color= 'red')
        plt.xlabel("AÑO", color='red')
        plt.ylabel("TWh", color='red');
        
    elif tipo == plt.bar: 
        plt.bar(x, y)
        plt.title("Consumo eolico", color= 'red')
        plt.xlabel("AÑO", color='red')
        plt.ylabel("TWh", color='red');

grafica_eolica('European Union (27)', plt.plot)
```

```python
###### PREPARAMOS LAS GRAFICAS PARA LA PRESENTACION
plt.figure(figsize=(7,10))#Se  inicializa la figura

plt.subplot(3, 1, 1) # filas, columnas, posición
grafica_eolica('World', plt.plot)
plt.title("CONSUMO EOLICO A NIVEL MUNDIAL", color= 'blue')
plt.xlabel("", color='red')
plt.ylabel("TWh", color='red');

plt.subplot(3, 1, 2) # filas, columnas, posición
grafica_eolica('European Union (27)', plt.plot)
grafica_eolica('Spain',plt.bar)
plt.title("CONSUMO EOLICO EN EUROPA   /    España en columnas", color= 'green')
plt.xlabel("", color='red')
plt.ylabel("TWh", color='red');

plt.subplot(3,1,3)
grafica_eolica('Africa',plt.plot)
grafica_eolica('South Africa', plt.plot)
plt.title("CONSUMO EOLICO EN AFRICA / SudAfrica naranja", color= 'black')
plt.xlabel("", color='red')
plt.ylabel("TWh", color='red');


```
